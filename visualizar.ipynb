````python
#ayudame a crear un codigo para visualizar archivos .pkl
#¡Claro! A continuación te proporciono un ejemplo de código en Python que te permitirá visualizar archivos `.pkl` (Pickle) utilizando la biblioteca `pickle` para cargar los datos y `pandas` para visualizarlos de manera más amigable si se trata de un DataFrame.```python


import pickle
import pandas as pd
import joblib
import matplotlib.pyplot as plt

# Cargar el archivo .pkl
with open(r'E:\Unach\Semestre4\ProyectoML&BDD\IF-Base-datos-Machine-learning\Models\Random_Forest\random_forest_model.pkl', 'rb') as f:
    try:
        data = pickle.load(f)
    except Exception:
        f.seek(0)
        data = joblib.load(f)

# Si los datos son un DataFrame, puedes visualizarlos directamente
if isinstance(data, pd.DataFrame):
    data.head()  # Muestra las primeras filas
    data.plot()  # Visualiza el DataFrame
    plt.show()
else:
    print("Los datos no son un DataFrame.")
````
